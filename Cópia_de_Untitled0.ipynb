```python
#converter para numerico
df['Conceito Enade (Faixa)'] = pd.to_numeric(df['Conceito Enade (Faixa)'], errors='coerce')

areas = df.groupby(['Área de Avaliação', 'Grau Acadêmico'])['Conceito Enade (Faixa)'].agg(
    ['mean', 'median', 'std', 'count']).sort_values('mean', ascending=False)

areas
df
```

```python
import matplotlib.pyplot as plt

# Agrupar as top/bottom 10 instituições
top = df[df['Conceito Enade (Faixa)'] == 5].groupby('Nome da IES*').size().nlargest(10)
bottom = df[df['Conceito Enade (Faixa)'] == 1].groupby('Nome da IES*').size().nlargest(10)

# Combinar em um único DataFrame
combinado = pd.DataFrame({
    'Cursos Nota 5': top,
    'Cursos Nota 1': bottom
}).fillna(0)  # Substitui NaN por 0 onde não houver valor

# Ordenar para facilitar leitura
combinado = combinado.sort_values(by='Cursos Nota 5', ascending=False)

# Plotar gráfico de barras agrupadas
ax = combinado.plot(kind='bar', figsize=(16, 6), color=['blue', 'red'])

plt.title('Top 10 instituições com mais cursos nota 5 e nota 1 (ENADE)')
plt.ylabel('Quantidade de cursos')
plt.xlabel('Instituições')
plt.xticks(rotation=70, ha='right')
plt.legend(title='Conceito ENADE')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Cálculo de estatísticas por UF
uf = df.groupby('Sigla da UF** ')['Conceito Enade (Faixa)'].agg(
    # calcula as principais estatísticas para o conceito ENADE por estado e depois organiza os resultados para que os estados com as maiores médias apareçam primeiro.
    ['mean', 'std', 'count']).sort_values('mean', ascending=False)

# Visualização: Boxplot
plt.figure(figsize=(14, 8))
sns.boxplot(data=df, x='Sigla da UF** ', y='Conceito Enade (Faixa)', showfliers=False)
plt.title('Distribuição de Conceitos por Estado', fontsize=16)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('conceito_por_uf_boxplot.png', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Garantir que uf está no formato correto para merge
uf_reset = uf.reset_index()
uf_reset = uf_reset.rename(columns={'Sigla da UF** ': 'abbrev_state'})

# Realizar o merge com o GeoDataFrame dos estados
brasil_com_dados = brasil.merge(uf_reset, on='abbrev_state', how='left')

# Reprojetar para um CRS projetado para  centroides correto
brasil_proj = brasil_com_dados.to_crs(epsg=3857)
centroides = brasil_proj.geometry.centroid

# projetar os centroides
centroides = centroides.to_crs(epsg=4326)

# Plotar o mapa coroplético
fig, ax = plt.subplots(figsize=(12, 10))
brasil_com_dados.plot(
    column='mean',
    cmap='viridis',
    linewidth=0.8,
    edgecolor='0.8',
    legend=True,
    ax=ax,
    missing_kwds={'color': 'lightgrey', 'label': 'Dados Ausentes'}
)

# Adicionar rótulos com as siglas dos estados nos centroides
for x, y, label in zip(centroides.x, centroides.y, brasil_com_dados['abbrev_state']):
    ax.text(x, y, label, fontsize=8, ha='center')

ax.set_title('Conceito ENADE Médio por Estado (2021)', fontsize=16)
ax.axis('off')
plt.tight_layout()
plt.savefig('mapa_conceito_enade_estado.png', dpi=300)
plt.show()
```

```python


# Ler dados geográficos e ENADE
brasil = read_state()
uf = df.groupby('Sigla da UF** ')['Conceito Enade (Faixa)'].agg(['mean', 'std', 'count'])
uf_reset = uf.reset_index().rename(columns={'Sigla da UF** ': 'abbrev_state'})
brasil_com_dados = brasil.merge(uf_reset, on='abbrev_state', how='left')

# Reprojetar para cálculo de centroides
brasil_proj = brasil_com_dados.to_crs(epsg=3857)
centroides = brasil_proj.geometry.centroid.to_crs(epsg=4326)

# Identificar melhor curso por estado
top_cursos = (
    df.groupby(['Sigla da UF** ', 'Área de Avaliação'])['Conceito Enade (Faixa)']
    .mean().reset_index()
    .sort_values(['Sigla da UF** ', 'Conceito Enade (Faixa)'], ascending=[True, False])
    .drop_duplicates('Sigla da UF** ')
    .rename(columns={'Sigla da UF** ': 'abbrev_state'})
)

# Indexar cursos com números
top_cursos['curso_id'] = range(1, len(top_cursos) + 1)

# Merge com geometria e calcular centroides
top_cursos = top_cursos.merge(brasil_proj[['abbrev_state', 'geometry']], on='abbrev_state', how='left')
top_cursos = gpd.GeoDataFrame(top_cursos, geometry='geometry', crs=brasil_proj.crs)
top_cursos['centroid'] = top_cursos.geometry.centroid.to_crs(epsg=4326)

# Plotar o mapa
fig, ax = plt.subplots(figsize=(12, 10))
brasil_com_dados.plot(
    column='mean',
    cmap='viridis',
    linewidth=0.8,
    edgecolor='0.8',
    legend=True,
    ax=ax,
    missing_kwds={'color': 'lightgrey', 'label': 'Dados Ausentes'}
)

# Adicionar siglas dos estados
for x, y, label in zip(centroides.x, centroides.y, brasil_com_dados['abbrev_state']):
    ax.text(x, y, label, fontsize=8, ha='center', color='white', weight='bold')

# Adicionar número do curso por estado
for idx, row in top_cursos.iterrows():
    x, y = row['centroid'].x, row['centroid'].y
    curso_idx = row['curso_id']
    ax.text(x, y - 0.5, str(curso_idx), fontsize=8, ha='center', color='black', weight='bold')

# Título e ajustes
ax.set_title('Conceito ENADE Médio por Estado (2021) + Curso Destaque', fontsize=16)
ax.axis('off')
plt.tight_layout()
plt.savefig('mapa_enade_legenda_numerada.png', dpi=300)
plt.show()

# Exibir a legenda dos cursos
print("Legenda dos Cursos por Estado:")
for _, row in top_cursos.sort_values('curso_id').iterrows():
    print(f"{row['curso_id']:>2} - {row['Área de Avaliação']} ({row['abbrev_state']})")
```
